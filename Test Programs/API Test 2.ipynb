```python
# join fixtures to teams

df = pd.merge(
    left=fixtures,
    right=teams,
    left_on='team_a',
    right_on='id'
)

df2 = pd.merge(
    left=df,
    right=teams,
    left_on='team_h',
    right_on='id'
)

df2.head(5)

```

```python
filtered_df = df2[ (df2['team_h'] == 15) | (df2['team_a'] == 15) ]
filtered_df = filtered_df.reset_index()
#filtered_df[['name_y','team_h_score', 'team_a_score', 'name_x']].head(11)

w = 0
d = 0
l = 0

for index, row in filtered_df.iterrows():
    # print(row['name_y'],row['team_h_score'], row['team_a_score'], row['name_x'])
    if row['team_h'] == 15 and (row['team_h_score'] > row['team_a_score']):
        w += 1
        # print("home win")
    elif row['team_h'] == 15 and (row['team_h_score'] == row['team_a_score']):
        d += 1
        # print("home draw")
    elif row['team_h'] == 15 and (row['team_h_score'] < row['team_a_score']):
        l += 1
        # print("home loss")

    if row['team_a'] == 15 and (row['team_a_score'] > row['team_h_score']):
        w += 1
        # print("away win")
    elif row['team_a'] == 15 and (row['team_a_score'] == row['team_h_score']):
        d += 1
        # print("away draw")
    elif row['team_a'] == 15 and (row['team_a_score'] < row['team_h_score']):
        l += 1
        # print("away loss")

print(f"NUFC Record: Played: {w+d+l}, Won: {w}, Draw: {d}, Loss: {l}, Points: {(w*3)+(d)}")



#print(df2.columns.tolist())
#print(df2.loc[ ( (df2['team_h'] == 15) | (df2['team_a'] == 15) ) ] )

#new_df = df2.where(((df2['team_h'] == 15) | (df2['team_a'] == 15)))
#new_df[['name_y','team_h_score', 'team_a_score', 'name_x']].head(11)
```

```python
def get_team_record(team_id):
    team_name = teams.loc[teams.id == team_id, 'name'].values[0]

    record_df = df2[ (df2['team_h'] == team_id) | (df2['team_a'] == team_id) ]
    record_df = record_df.reset_index()
    #record_df[['name_y','team_h_score', 'team_a_score', 'name_x']].head(11)

    w = 0
    d = 0
    l = 0

    gd = 0

    for index, row in record_df.iterrows():
        # print(row['name_y'],row['team_h_score'], row['team_a_score'], row['name_x'])
        if row['team_h'] == team_id and (row['team_h_score'] > row['team_a_score']):
            w += 1
            # print("home win")
        elif row['team_h'] == team_id and (row['team_h_score'] == row['team_a_score']):
            d += 1
            # print("home draw")
        elif row['team_h'] == team_id and (row['team_h_score'] < row['team_a_score']):
            l += 1
            # print("home loss")

        if row['team_a'] == team_id and (row['team_a_score'] > row['team_h_score']):
            w += 1
            # print("away win")
        elif row['team_a'] == team_id and (row['team_a_score'] == row['team_h_score']):
            d += 1
            # print("away draw")
        elif row['team_a'] == team_id and (row['team_a_score'] < row['team_h_score']):
            l += 1
            # print("away loss")

    for index, row in record_df.iterrows():
        # if team is at home, add home goals and subtract away goals
        if row['team_h'] == team_id and (row['team_h_score'] == row['team_h_score']):
            gd += row['team_h_score']
            gd -= row['team_a_score']
        # if team is away, subtract home goals and add away goals
        elif row['team_a'] == team_id and (row['team_h_score'] == row['team_h_score']):
            gd -= row['team_h_score']
            gd += row['team_a_score']

    gd = int(gd)
    
    remaining_df = df2[ (df2['finished'] == False) & ((df2['team_h'] == team_id) | (df2['team_a'] == team_id)) ]
    remaining_df = remaining_df.reset_index()
    left = int(len(remaining_df))
    
    
    pts = int((w*3)+(d))
    ply = int(w+d+l)

    max = int(pts + 3*left)

    print(f"{team_name} Record: Played: {ply}, Won: {w}, Draw: {d}, Loss: {l}, Points: {pts}, Goal Difference: {gd}, Remaining: {left}")

    return pts, ply, left, max, gd

points, played, remaining, max, gd = get_team_record(16)

```
